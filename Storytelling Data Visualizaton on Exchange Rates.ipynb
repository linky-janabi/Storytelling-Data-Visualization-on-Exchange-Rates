## Guided Project: Storytelling Data Visualization on Exchange Rates

The dataset we'll use describes Euro daily exchange rates between 1999 and 2021. The euro (symbolized with €) is the official currency in most of the countries of the European Union. The objective of this analysis is to visualize and explain the trend of the euro versus USD exchange rates over a period of time using two points of reference i.e.
- exchange rate trends under the reign of the last 3 US presidents
- exchange rate trends during the 2007-2008 financial crisis.

```python
# Reading in the euro-daily-hist_1999_2020.csv dataset using pandas

import pandas as pd

exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

- columns with float and string data types

### Data Cleaning

Our focus in this project will be on the exchange rate between the **Euro** and the **American dollar.**

```python
#renaming [US dollar] and Period/Unit: columns

exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)

#Changing Time column to a datetime datatype, 
#sort values and reset the index

exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head()
```

```python
#isolating the Time and US_dollar columns

euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts() #'-' there are 62 characters
```

```python
#dropping all the '-' characters
#converting the US_dollar column to float

euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

### Rolling Mean

We want to find the US dollar rate against euro change from time to time. 
- Short term trends (visual representation of daily variation in the exchange rate)
- Long term trends (visual representation of rolling mean in the exchange rate).

```python
import matplotlib.pyplot as plt
%matplotlib inline

#visual representation of daily variation in the exchage rate

plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])

plt.show()
```

```python
#long-term visualization of exchange rate rolling mean
#using rolling window of 7,30,50,100,365

plt.figure(figsize=(9,6))

plt.subplot(3,2,1)

plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original Values', weight='bold')

for i, rolling_mean in zip([2,3,4,5,6], [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')

plt.tight_layout()
plt.show()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

### EURO-USD Exchange Rates Under Three US Presidents

We are going to do an explanatory data visualization of the exchange rates under the reign of the last three US presidents i.e. George Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)

```python
#isolating the presidential periods respectively

bush_obama_trump = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >=2001) & (euro_to_dollar['Time'].dt.year <2021)]
bush = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year < 2009)]
obama= bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >=2009) & (bush_obama_trump['Time'].dt.year < 2017)]
trump= bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >=2017) & (bush_obama_trump['Time'].dt.year < 2021)]
```

Below, we will use matplotlib's functional approach to build the graphs. We use this approach because it offers more flexibility in arranging the subplots.

```python
#we will use the fivethrity style

import matplotlib.style as style
style.use('fivethirtyeight')

#build the subplots using matplotlib's functional approach

plt.figure(figsize=(12,6))
ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
ax4 = plt.subplot(2,1,2)
axes = [ax1, ax2, ax3, ax4]

for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2', '1.4', '1.6'], alpha=0.3)
    ax.grid(alpha=0.5)

#plot ax1:Bush

ax1.plot(bush['Time'], bush['rolling_mean'], color='#3357FF')
ax1.set_xticklabels(['2001','' , '2003', '', '2005', '','2007', '', '2009'], alpha=0.3)
ax1.text(731516.0, 1.92, 'Bush', fontsize=18, weight='bold', color='#3357FF')
ax1.text(731216.0, 1.8, '(2001-2009)', weight='bold', alpha=0.3)

#plot ax2:Obama

ax2.plot(obama['Time'], obama['rolling_mean'], color='#FFC300')
ax2.set_xticklabels(['2001', '', '2003', '', '2005', '', '2007', '', '2009'], alpha=0.3)
ax2.text(734288.0, 1.92, 'Obama', fontsize=18, weight='bold', color='#FFC300')
ax2.text(734138.0, 1.8, '(2009-2017)', weight='bold', alpha=0.3)

#plot ax3:Trump

ax3.plot(trump['Time'], trump['rolling_mean'], color='#900C3F')
ax3.set_xticklabels(['2001', '', '2003', '', '2005', '', '2007', '', '2009'], alpha=0.3)
ax3.text(736855.0, 1.92, 'Trump', fontsize=18, weight='bold', color='#900C3F')
ax3.text(736745.0, 1.8, '(2017-2021)', weight='bold', alpha=0.3)

#plotax4:Bush, Obama, Trump

ax4.plot(bush['Time'], bush['rolling_mean'], color='#3357FF')
ax4.plot(obama['Time'], obama['rolling_mean'], color='#FFC300')
ax4.plot(trump['Time'], trump['rolling_mean'], color='#900C3F')
ax4.set_xticks([])
ax4.grid(alpha=0.3)

#Adding signature
ax4.text(729916.0, 0.65, '©DATAQUEST' + ' '*103 + 'Source: European Central Bank',
        color='#f0f0f0', backgroundcolor='#4d4d4d',
        size=14)

#Adding title
ax1.text(730016.0, 2.35, 'EURO-USD averaged at 1.22 under the last three US presidents', fontsize=20, weight='bold')
ax1.text(730016.0, 2.14, '''EURO-USD exchage rates witnessed during the reign of presidents; George Bush(2001-2009), Barrack Obama(2009-2017) and Donald Trump(2017-2021)''', fontsize=16)

plt.show()
```

### EURO-USD Exchange Rate During the Financial Crisis

We will visualize the rates graphically to explain the trend witnessed during the financial crisis in 2007-2008 period. We will also show the data for 2006 and 2009 for comparison.

```python
#isolating the financial crisis period i.e. 2007-2008
#isolating 2006-2009 periods for comparative analysis

fin_crisis_7_8 = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2007) & (euro_to_dollar['Time'].dt.year <= 2008)]
fin_crisis_6_9 = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2006) & (euro_to_dollar['Time'].dt.year <= 2009)]
```

```python
fig, ax = plt.subplots(figsize=(11, 3.5))

#adding the plots
ax.plot(fin_crisis_6_9['Time'], fin_crisis_6_9['rolling_mean'], linewidth=1, color='#3357FF')
ax.plot(fin_crisis_7_8['Time'], fin_crisis_7_8['rolling_mean'], linewidth=1, color='#FF33A1')

#highliting the peak of the crisis using a vertical span

ax.axvspan(xmin=733112.0, xmax=733302.0, ymin=0.09,
           alpha=0.3, color='grey')

#customizing the tick labels

ax.set_xticklabels([])
ax.set_yticklabels([])
x= 732272.0

for year in ['2006', '2007', '2008', '2009']:
    ax.text(x, 1.13, year, alpha=0.5, fontsize=10)
    x+= 365
    
y= 1.193

for rates in ['1.2','1.3','1.4','1.5']:
    ax.text(732172.0, y, rates, alpha=0.5, fontsize=10)
    y+=0.1

#Adding title and subtitle
ax.text(732172.0, 1.67, 'EURO-USD hit a record high of 1.59 during the financial crisis between 2007-2008', fontsize=20, weight='bold')
ax.text(732172.0, 1.63, 'Euro-Usd exchange rates between 2006-2010', fontsize=16)
ax.grid(alpha=0.3)

#Adding a signature
ax.text(732172.0, 1.07, '@DATAQUEST' + ' '*103 + 'Source: European Central Bank',
        color='#f0f0f0', backgroundcolor='#4d4d4d',
        size=14 )

plt.show()
```
